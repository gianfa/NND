```python
%matplotlib inline
import sys
sys.path.append('../')
from dnn_designer.funx import *
import matplotlib.pyplot as plt
import numpy as np
```

Basic function `compute_parlpd`.  
Computes a parallelepiped.

```python
# prepare some coordinates
X, Y, Z = np.indices((20, 20, 20)) # axes length 0->20
axes = (X, Y, Z)

# define cuboid
dims_1 = (1,1,1) # dx, dy, dz
pos_1 = (5,5,0) # pos x, pos y, pos z
cube1 = compute_parlpd(dims_1, pos_1, axes)


# draw it
# 1. draw the scene
voxelarray =  cube1 # combines all the shapes in the scene. in this case only cube1
# 2. create the color array for each shape
colors = np.empty(voxelarray.shape, dtype=object) # 
colors[cube1] = 'blue'
# 3. launch the plot
ax = plt.figure(figsize=(6,6)).add_subplot(projection='3d')
ax.voxels(voxelarray, facecolors=colors, edgecolor='k')
plt.show()
```

Function to draw a Dense layer, `draw_Dense`

```python
# prepare some coordinates
X, Y, Z = np.indices((20, 20, 20)) # axes length 0->20
axes = (X, Y, Z)

# define cuboid
n = 10 # <-- only the number of units is needed
pos_1 = (5,5,0) # pos x, pos y, pos z
dense = draw_Dense(n, pos_1, axes)


# draw it
# 1. draw the scene
voxelarray =  dense # combines all the shapes in the scene. in this case only dense
# 2. create the color array for each shape
colors = np.empty(voxelarray.shape, dtype=object) # 
colors[dense] = 'blue'
# 3. launch the plot
ax = plt.figure(figsize=(6,6)).add_subplot(projection='3d')
ax.voxels(voxelarray, facecolors=colors, edgecolor='k')
plt.show()
```

Function to draw a $n \times m$ 2D Dense layer, `draw_Dense2D`.

```python
# prepare some coordinates
X, Y, Z = np.indices((20, 20, 20)) # axes length 0->20
axes = (X, Y, Z)

# define cuboid
n, m = 5,5
pos_1 = (5,5,0) # pos x, pos y, pos z
dense = draw_Dense2D(n, m, pos_1, axes)


# draw it
# 1. draw the scene
voxelarray =  dense # combines all the shapes in the scene. in this case only dense
# 2. create the color array for each shape
colors = np.empty(voxelarray.shape, dtype=object) # 
colors[dense] = 'blue'
# 3. launch the plot
ax = plt.figure(figsize=(6,6)).add_subplot(projection='3d')
ax.voxels(voxelarray, facecolors=colors, edgecolor='k')
plt.show()
```

Example

```python
# prepare some coordinates #
X_len = 20; Y_len=20; Z_len=20
X, Y, Z = np.indices((X_len, Y_len, Z_len)) # axes length 0->20
axes = (X, Y, Z)

# draw shapes #
dims_1 = (1,1,1) # dx, dy, dz
pos_1 = (14,3.5,0)
cube1 = compute_parlpd(dims_1, pos_1, axes)

pos_2 = (3,0,0)
dense1 = draw_Dense(10, pos_2, axes)

pos_3 = (6,0,0)
dense2d_1 = draw_Dense2D(n=10, m=10, pos=pos_3, axes=axes)

pos_4 = (8,0,0)
dense2d_2 = draw_Dense2D(n=10, m=10, pos=pos_4, axes=axes)

pos_5 = (10,2.5,0)
dense2d_3 = draw_Dense2D(n=4, m=4, pos=pos_5, axes=axes)

# combine the objects into a single boolean array
# combine_shapes()
voxelarray =  cube1 | dense2d_1 | dense2d_2 | dense2d_3

# set the colors of each object
# to move above into a World lass
colors = np.empty(voxelarray.shape, dtype=object)
colors[cube1] = 'blue'
colors[dense2d_1] = 'blue'
colors[dense2d_2] = 'orange'
colors[dense2d_3] = 'orange'

## Start the plot ##
ax = plt.figure(figsize=(10,10)).add_subplot(projection='3d')
ax.voxels(voxelarray, facecolors=colors, edgecolor='k')

# add text
model_name = 'My model'
# fig.suptitle(model_name, fontsize=20, y=0.1)
ax.text(2, Y_len/2, Z_len*0.9,r'My model', fontsize=22)

ax.text(pos_2[0], -1, 0,r'Input(Dense(10x10))', fontsize=11)
ax.text(pos_3[0]+2, -5.7, 3,r'Dense(10x10)', fontsize=11)
ax.text(pos_4[0]+4, -3, 3,r'Dense(5x5)', fontsize=11)
ax.text(pos_1[0]+1, -3.3, 3,r'Output:Dense(1)', fontsize=11)

# Rotate the plot 3D
angle = -72
ax.view_init(30, angle)

#  Remove plot grid and axis
plt.grid(False)
plt.axis('off')
plt.show()
```
